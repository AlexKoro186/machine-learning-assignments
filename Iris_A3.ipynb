```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# 1. Datensatz laden
iris = load_iris()
X = iris.data
y = iris.target

# 2. Aufteilen in 60% Training, 40% Test (reproduzierbar)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.4, random_state=42, shuffle=True
)

# 3. KNN-Modell mit 3 Nachbarn
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)
acc_knn = accuracy_score(y_test, y_pred_knn)

# 4. Naive Bayes (GaussianNB)
nb = GaussianNB()
nb.fit(X_train, y_train)
y_pred_nb = nb.predict(X_test)
acc_nb = accuracy_score(y_test, y_pred_nb)

# 5. Ergebnisse vergleichen
print("Genauigkeit KNN (k=3): {:.2f}%".format(acc_knn * 100))
print("Genauigkeit Naive Bayes (GaussianNB): {:.2f}%".format(acc_nb * 100))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# 1. Datensatz laden
iris = load_iris()
X = iris.data
y = iris.target

# 2. Aufteilen in 60% Training, 40% Test (reproduzierbar)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.4, random_state=42, shuffle=True
)

model = DecisionTreeClassifier(
    max_depth=5,
    min_samples_leaf=5,
    min_impurity_decrease=0.1,
    random_state=42
)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```
