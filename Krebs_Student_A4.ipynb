```python
# Gegebene Werte
TP = 542
FP = 108
FN = 650
TN = 98700

# Berechnungen
accuracy    = (TP + TN) / (TP + TN + FP + FN)
precision   = TP / (TP + FP)
recall      = TP / (TP + FN)          # Sensitivität
specificity = TN / (TN + FP)
f1_score    = 2 * (precision * recall) / (precision + recall)

# Ausgabe
print("=== Modellmetriken ===")
print(f"Accuracy:    {accuracy:.4f} ({accuracy*100:.2f}%)")
print(f"Precision:   {precision:.4f} ({precision*100:.2f}%)")
print(f"Recall:      {recall:.4f} ({recall*100:.2f}%)")
print(f"Specificity: {specificity:.4f} ({specificity*100:.2f}%)")
print(f"F1-Score:    {f1_score:.4f} ({f1_score*100:.2f}%)")
```

```python
import math

# Beispielkoeffizienten
b0 = -10.65
b1 = 0.0055       # Saldo
b2 = -0.15        # Einkommen (in 10.000 USD)
b3 = -0.8         # Student (1 = ja)

# Eingabewerte
saldo = 1500
income = 40000
student = 1  # 1 = Student, 0 = kein Student

# lineare Kombination
z = b0 + b1 * saldo + b2 * (income / 10000) + b3 * student

# Wahrscheinlichkeit
p = 1 / (1 + math.exp(-z))

print(f"Ausfallwahrscheinlichkeit (Student): {p:.4f} ({p*100:.2f}%)")

# Kein Student
student = 0
z_no = b0 + b1 * saldo + b2 * (income / 10000) + b3 * student
p_no = 1 / (1 + math.exp(-z_no))

print(f"Ausfallwahrscheinlichkeit (kein Student): {p_no:.4f} ({p_no*100:.2f}%)")
```
